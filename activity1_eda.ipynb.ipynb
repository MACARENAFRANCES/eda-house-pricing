```python
# Machine Learning & Artificial Intelligence Techniques
# Economics Degree
# Exploratory Data Analysis (EDA)
#
# Academic project
# Uploaded to GitHub: 03 November 2025

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import plotly.express as px

# Dataset loading
# (Colab upload omitted – local execution in JupyterLab)
data = pd.read_csv("house_pricing_train_Actividad1.csv", sep=",")

# First look at the dataset
data.head()

# Dataset dimensions (instances and variables)
data.shape

# Descriptive statistics of the dataset
data.describe()

# Descriptive statistics of the target variable
data["SalePrice"].describe()

# Dataset structure and variable types
data.info()

# The dataset contains both numerical and categorical variables related to
# housing characteristics, construction quality, size, and sale conditions.
# SalePrice shows a right-skewed distribution with higher-price outliers.

# Check unique values of HeatingQC
print("HeatingQC values:", data["HeatingQC"].unique())

# Histograms of numerical variables
data.hist(figsize=(16, 20), bins=50, xlabelsize=8, ylabelsize=8);

# Histogram of SalePrice
plt.figure(figsize=(12, 10))
plt.hist(data["SalePrice"], bins=25)
plt.title("Distribution of SalePrice")
plt.xlabel("SalePrice")
plt.ylabel("Frequency")
plt.show()

# Missing values analysis
data.isnull().sum()

# Some variables contain missing values, mainly related to optional
# housing features. Certain categorical nulls are interpreted as 'None'.

# Ratio of missing values
data_nulls = data.isnull().sum().sort_values(ascending=False)
data_nulls_ratio = (data_nulls / len(data)).reset_index()
data_nulls_ratio.columns = ["Feature", "ratio_nulls"]
data_nulls_ratio[data_nulls_ratio["ratio_nulls"] > 0]

# Fill missing values for MasVnrType
data["MasVnrType"] = data["MasVnrType"].fillna("None")

# Recheck missing values
data.isnull().sum().sort_values(ascending=False)

# Drop Id column (identifier, not useful for analysis)
data = data.drop("Id", axis=1)

data.head(10)

# Boxplot for outlier detection
plt.figure(figsize=(15, 8))
sns.boxplot(data=data)
plt.xticks(rotation=45)
plt.title("Boxplot for Outlier Detection")
plt.show()

# Histograms and boxplots show skewed distributions and the presence of outliers.
# These extreme values correspond to houses with very particular characteristics.

# Correlation analysis
numeric_data = data.select_dtypes(include=["float64", "int64"])
correlation_matrix = numeric_data.corr(method="pearson")

# Heatmap of correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Matrix")
plt.show()

# The correlation matrix highlights strong positive correlations between
# SalePrice and variables related to size and quality of the property.
# These variables are particularly relevant for predictive modeling.
```
